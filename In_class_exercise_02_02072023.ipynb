```python
# You code here (Please add comments in the code):
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
import json

movies_url = "https://www.imdb.com/title/tt6751668/reviews?ref_=tt_urv"
pages = urllib.request.urlopen(movies_url)
soup = BeautifulSoup(pages)

UserName = []
star = []
review_title = []
review_text = []
review_posted_time = []

for user in soup.find_all("span",{'class':'display-name-link'}, 'a'):
  user = UserName.append(user.text.strip())

for Rating in soup.find_all("span",{'class':'rating-other-user-rating'},'span'):
  Rating = star.append(Rating.text.strip())

for title in soup.find_all("a",{'class':'title'}):
  title = review_title.append(title.text.strip())

for reviewtext in soup.find_all("div", {'class':'text show-more__control'}):
  reviewtext = review_text.append(reviewtext.text.strip())

for postedtime in soup.find_all("span",{'class':'review-date'}):
  postedtime = review_posted_time.append(postedtime.text.strip())

rev = {'User Name':UserName, 'Ratings':star,'Review title':review_title, 'Review':review_text, 'Review Posted Time':review_posted_time}
  
reviews_1 = pd.DataFrame.from_dict(rev, orient='index')
reviews_1.transpose()
```

Question 3 (10 points): Write python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "information retrieval". The articles should be published in the last 10 years (2012-2022).

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
# You code here (Please add comments in the code):
import requests
query_params = {
    'query' : 'information retrieval',
    'offset' : 0,
    'limit' : 100,
    'fields' : 'title,authors,abstract,year,venue',
    'year' : '2012-2022'
}
url = "http://api.semanticscholar.org/graph/v1/paper/search?query=information+retrieval"
# response = requests.get(url, params=query_params)

all_papers = []
for i in range(10):
  query_params['offset'] += 10
  response = requests.get(url, params=query_params)
  all_papers.extend(response.json()['data'])

import pandas as pd
filtered_list = []
for paper in all_papers:
  p = {}
  p['title'] = paper['title']
  p['venue'] = paper['venue']
  p['year'] = paper['year']
  p['authors'] = paper['authors']
  p['abstract'] = paper['abstract']

  filtered_list.append(p)

data_frame = pd.DataFrame.from_dict(filtered_list)
data_frame
```
